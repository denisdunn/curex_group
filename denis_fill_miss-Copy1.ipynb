ReceiveTime	The time the order was received (UTC)

ActivationTime	The time the order was activated in the matching engine (UTC)

ReceiveTimeEST	The time the order was received (EST)

ActivationTimeEST	The time the order was activated in the matching engine (EST)

Year	The trading year

WeekOfYear	The trading week number (1 - 52)

TradingWeek	The starting date of the trading week

TradeDate	The trade date

MissClass	Order misses bucketed by timeframes in milliseconds

MissDelta	Order missed milliseconds

OrderId	The order id

Symbol	The currency pair symbol

OrderType	The order type (limit, peg, twap, vwap)

TIF	Time in force (GFD, DAY, GTC)

BuySell	BUY or SELL

MissFill	The order is classified as a miss or a fill

MissFillPartial	The order is classified as miss, fill or partial

OrderPrice	The limit price submitted on the order

OrderAmount	The order amount

OrderAmountDone	The order amount filled

USDAmountDone	The order amount done in USD

USDAmountLeft	The order amount unfilled in USD

PlacementBid	The top of book bid at the time the order was received

PlacementAsk	The top of book offer at the time the order was received

ActivationBid	The top of book bid at the time the order was activated in the matching engine

ActivationAsk	The top of book offer at the time the order was activated in the matching engine

MATCH_EXISTED_PRIOR_TO_PLACEMENT	Indicates if a match existed prior to order placement

MATCH_EXISTED_AT_PLACEMENT_TIME	Indicates if a match existed at order placement

MATCH_EXISTED_DURING_RANDOM_DELAY	Indicates if a match existed while in the random delay

PriceExpirationTime	If there was a matching price, the timestamp of when the prices expired

MissReason	The reason for a miss:  FILLED - the order was filled, NO PRICE - there was no matching 
price,

PRICE EXPIRED PRIOR TO PLACEMENT - the price expired before receiving the order,

PRICE EXPIRED IN RANDOM DELAY - the price expired before the order was activated in the matching engin.

USDEquivalentAmount	The order amount in USD

USDConverationPrice	The order amount USD conversion rate

OrderPlacementPosition	The order placement relative book position (INSIDE, JOINED, AWAY, MATCHED)

OrderActivationPosition	The order activation relative book position (INSIDE, JOINED, AWAY, MATCHED)

MissPriceDeltaPips	The difference in pips between the order price and the top of book

PlacementSpread	The top of book spread at order activation time

Vol Fill Pct	Filled volume percentage

Vol Unfilled Pct	Unfilled volume percentage

```python
def normalize_by_group(df, by):
    groups = df.groupby(by)
    # computes group-wise mean/std,
    # then auto broadcasts to size of group chunk
    mean = groups.transform(np.mean)
    std = groups.transform(np.std)
    return (df[mean.columns] - mean) / std
```
